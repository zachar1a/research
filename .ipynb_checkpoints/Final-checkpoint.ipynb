```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_americas = pd.read_json('JSON Match Data/americasMatchData.txt')
df_europe   = pd.read_json('JSON Match Data/europeMatchData.txt')
df_asia     = pd.read_json('JSON Match Data/asiaMatchData.txt')
```

```python
info_americas = pd.json_normalize(df_americas['info'])
info_europe   = pd.json_normalize(df_europe['info'])
info_asia     = pd.json_normalize(df_asia['info'])
```

```python
info_americas = info_americas[info_americas.gameMode == 'CLASSIC'].reset_index()
info_americas.head()
```

```python
info_europe = info_europe[info_europe.gameMode == 'CLASSIC'].reset_index()
info_europe.head()
```

```python
info_asia = info_asia[info_asia.gameMode == 'CLASSIC'].reset_index()
info_asia.head()
```

```python
#participants = pd.json_normalize(info['participants'][0])
#tmp = pd.json_normalize(info['participants'][1])
#participants = participants.append(tmp, ignore_index=True)
#participants = pd.DataFrame()
participants_americas = pd.json_normalize(info_americas['participants'][0])
for players in info_americas['participants'][1:10000]:
    tmp_americas = pd.json_normalize(players)
    participants_americas = participants_americas.append(tmp_americas, ignore_index=True)

participants_europe = pd.json_normalize(info_europe['participants'][0])
for players in info_europe['participants'][1:10000]:
    tmp_europe = pd.json_normalize(players)
    participants_europe = participants_europe.append(tmp_europe, ignore_index=True)
    
participants_asia = pd.json_normalize(info_asia['participants'][0])
for players in info_europe['participants'][1:10000]:
    tmp_asia = pd.json_normalize(players)
    participants_asia = participants_asia.append(tmp_asia, ignore_index=True)
```

```python
x = participants_americas[['champLevel', 'kills']]

# If this doesn't work then i need to go back to 
# iteratively appending 1 or 0 to list
y = list(participants_americas['win']*1)
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y, marker='o',linestyle='None')
```

```python
jungle  = 0
adc     = 0
top     = 0
support = 0
mid     = 0

#roles = [jungle, adc, top, support, mid]
for player in range(len(participants_americas)):
    if participants_americas['lane'][player] == 'JUNGLE':
        jungle += participants_americas['win'][player]
    elif participants_americas['lane'][player] == 'BOTTOM':
        if participants_americas['kills'][player] > 4:
            adc+=participants_americas['win'][player]
        else:
            support += participants_americas['win'][player]
    elif participants_americas['lane'][player] == 'TOP':
        top += participants_americas['win'][player]
    elif participants_americas['lane'][player] == 'MIDDLE':
        mid += participants_americas['win'][player]
roles = [jungle, adc, top, support, mid]
```

```python
lanes = ['Jungle', 'ADC', 'Top', 'Support', 'Mid']
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(lanes,roles)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
# LOGISTIC REGRESSION FOR K/D/A
regions = [participants_americas, participants_asia, participants_europe]
for i in range(len(regions)):
    region = regions[i]
    x = region[['kills','deaths','assists']]
    y = region['win']
    lr = LogisticRegression()
    lr.fit(x,y)
    if i == 0:
        print("Americas score {0}".format(lr.score(x,y)))
    if i == 1:
        print("Asia score {0}".format(lr.score(x,y)))
    if i == 2:
        print("Europe score {0}".format(lr.score(x,y)))
```

```python
for i in range(len(regions)):
    
    region = regions[i]
    x = region[['goldEarned','goldSpent']]
    y = region['win']
    lr = LogisticRegression()
    lr.fit(x,y)
    if i == 0:
        print("Americas score {0}".format(lr.score(x,y)))
    if i == 1:
        print("Asia score {0}".format(lr.score(x,y)))
    if i == 2:
        print("Europe score {0}".format(lr.score(x,y)))
```
